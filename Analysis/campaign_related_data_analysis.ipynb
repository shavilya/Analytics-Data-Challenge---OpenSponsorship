```python
# Import all the required libraries for data analysis 

import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt 
import seaborn as sns 

# For Warnings 
import warnings 
warnings.filterwarnings('ignore')

# For Display options 
pd.set_option('display.max_rows' , None)
pd.set_option('display.max_columns' , None)
```

```python
campaign_data_path = "../Data/sample_campaigns.xlsx"

campaign_data = pd.read_excel(campaign_data_path , sheet_name = 'data')
campaign_data.head()
```

```python
campaign_data.tail()
```

```python
campaign_data.info()
```

```python
campaign_data_df = campaign_data.copy()


campaign_data_df['published'] = campaign_data_df['published'].astype(object).where(campaign_data_df['published'].notnull() , 'Not Published' ) 
campaign_data_df.head()
```

```python
total_published_campaigns = len(campaign_data_df[campaign_data_df['published'] != 'Not Published'])
total_unpublished_campaigns = len(campaign_data_df[campaign_data_df['published'] == 'Not Published'])
total_brands = len(campaign_data_df['published'])

published_camp_ratio = (total_published_campaigns / total_brands) * 100  

print(f"Ratio of Brands with Published campaign : {round(published_camp_ratio,2)} %")
```
